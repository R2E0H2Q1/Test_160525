```python
print('\n***** Data Cleaning *****\n')

print('\n In the items table, there are products that appear multiple times (with the same name) and it'
'was decided to keep products with the same name, but keep the product with the most nonnull fields, and then delete the rest.'
'Note: (Follow the example shown in the class and implement it here).\n')

items_cleaning_df = pd.read_csv('https://raw.githubusercontent.com/pythonai250824/14.05.2025/refs/heads/main/Items.csv')
display(items_cleaning_df)

print('\n-----> First step - reserach: How many times an item appears on the table\n')

items_chk = items_cleaning_df.groupby('item_name')['id'].count()
display(items_chk[items_chk > 1])

print('\nSecond step  - see duplications\n')

dupitems = items_cleaning_df[items_cleaning_df['item_name'].duplicated(keep=False)].copy()
display(dupitems)

print('\n-----> Third step  - count not-null + sort by best first: Counts if all the cells are full per column, adds a new column with the count result\n')

dupitems['non_nan_count'] = dupitems.notna().sum(axis=1)
dupitems.sort_values(['item_name', 'non_nan_count'], ascending=[True, False], inplace=True)
display(dupitems)

print('\n-----> Fourth step  - count not-null + sort by best first:  From the dup items with the same id, keeps only the first\n')

kept = dupitems.drop_duplicates(subset=['item_name'], keep='first')
removed = dupitems[~dupitems['id'].isin(kept['id'])]
display(removed)

print('\n-----> Fifth step  - remove dup from books: It removes the duplicated items\n')

after_remove = items_cleaning_df[~items_cleaning_df['id'].isin(removed['id'])]
display(after_remove)

print('\n-----> Check: verify no more duplicates in after_remove\n')

items_chk = after_remove.groupby('item_name')['id'].count()
display(items_chk[items_chk > 1])  # Result needs to be empty 
```

```python
print('\n***** Method of Operation *****\n')

metop_df = pd.read_csv('https://raw.githubusercontent.com/pythonai250824/14.05.2025/refs/heads/main/Items.csv')

print('\n-----> 1. Add a new column counting how many non-null values exist in each row\n')

metop_df['non_null_count'] = metop_df.notna().sum(axis=1)
display(metop_df)

print('\n-----> 2. Sort by item name + the new column created, from largest to smallest\n')

sorted_items = metop_df.sort_values(['item_name', 'non_null_count'], ascending=[True, False])
display(sorted_items)

print('\n-----> 3. Remove duplicates by product name (keep only the first occurrence\n')

keeping = sorted_items.drop_duplicates(subset=['item_name'], keep='first').reset_index(drop=True)
removed = sorted_items[~sorted_items['id'].isin(keeping['id'])]
display(keeping)
```

```python
print('\n***** Updating the order_item table ****\n')

Updt_tbl = pd.read_csv('https://raw.githubusercontent.com/pythonai250824/14.05.2025/refs/heads/main/order_item.csv')
# display(Updt_tbl)

print('\n-----> 1. Replace the product IDs in the order_item table with the ID of the product that has the most non-null fields.\n')

beforeremvd = Updt_tbl.merge(items_cleaning_df[['id', 'item_name']], how='left', left_on='item_id', right_on='id')
display(beforeremvd)

afteremvd = beforeremvd.merge(after_remove[['id', 'item_name']], how='left', on='item_name')
display(afteremvd)

afteremvd.rename(columns={'id': 'new_id'}, inplace=True)

afteremvd['item_id'] = afteremvd['new_id']
display(afteremvd)
```

```python
print('\n Question 2 | Data Analysis |'
      'You have the following data files: Customers.csv, Items.csv, order_item.csv, orders.csv\n')

print('----->  3. Add a total_price column to the order_item table, calculating the purchase cost (product price * quantity):\n')

mergecheck_df = order_item_df.merge(items_df[['id', 'item_price']], left_on='item_id', right_on='id', how='left')
# display(mergecheck_df)
price_calc = mergecheck_df['item_price'] * mergecheck_df['quantity']
order_item_df['total_price'] = price_calc
display(order_item_df)

print('\n-----> 4. Display each purchase and its total price.\n')

totals = order_item_df[['order_id', 'total_price']]
display(totals)

print('\n-----> 5. Find what is the most expensive purchase? The cheapest? The average?\n')

display(Markdown(f"5.a. The biggest purchase is: {totals['total_price'].max()}"))
display(Markdown(f"5.b. The smallest purchase is: {totals['total_price'].min()}"))
display(Markdown(f"5.c. The average purchase is: {totals['total_price'].mean()}"))

print('\n-----> 6. From the customers table, create a PIVOT table and display each person by index, and gender as columns.'
      'Show how many customers belong to each gender within that nationality.\n')

customers_pivot = pd.pivot_table(customers_df, values='id', index='nationallity', columns='gender', aggfunc='count')
customers_pivot.fillna(0, inplace=True)
display(customers_pivot)
```
